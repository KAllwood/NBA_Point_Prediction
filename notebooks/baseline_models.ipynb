```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/preprocessed.csv")
```

```python
counts = df["PLAYER_ID"].value_counts()

valid_players = counts[counts >= 10].index

df = df[df["PLAYER_ID"].isin(valid_players)]
```

```python
def chronological_split(df, train_size):

    train_parts = []
    test_parts = []

    for _, player in df.groupby("PLAYER_ID"):

        split = int(len(player) * train_size)

        train_parts.append(player.iloc[:split])
        test_parts.append(player.iloc[split:])

    train = pd.concat(train_parts, ignore_index=True)
    test = pd.concat(test_parts, ignore_index=True)

    return train, test
```

```python
train_set , test_set = chronological_split(df,0.8)
```

```python
train_set[train_set['PLAYER_ID'] == 2544]
test_set[test_set['PLAYER_ID'] == 2544]
```

Baseline Model 1: The Season Average

```python
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, r2_score
train_mean = train_set["PTS"].mean()

predictions = np.full(
    len(test_set),
    train_mean
)

rmse = root_mean_squared_error(
    test_set["PTS"],
    predictions
)

mae=mean_absolute_error(test_set["PTS"],
    predictions
)



print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")

```

```python
player_means = (
    train_set.groupby("PLAYER_ID")["PTS"]
         .mean()
)
test_set["PLAYER_MEAN_PRED"] = test_set["PLAYER_ID"].map(player_means)

rmse = root_mean_squared_error(y_true= test_set['PTS'] , y_pred= test_set['PLAYER_MEAN_PRED'])

mae = mean_absolute_error(y_true= test_set['PTS'] , y_pred= test_set['PLAYER_MEAN_PRED'])

print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
```
